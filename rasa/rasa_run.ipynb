```python
!pip install -q rasa pyngrok

import os
import subprocess
import time
from pyngrok import ngrok

NGROK_AUTH = os.getenv("NGROK_AUTH_TOKEN")
if not NGROK_AUTH:
    from getpass import getpass
    NGROK_AUTH = getpass("Enter your ngrok auth token: ")
ngrok.set_auth_token(NGROK_AUTH)
```

```python
print("Starting Action Server...")
# Redirecting to file to avoid pipe buffering issues and allow background execution
with open("action_server.log", "w") as log_file:
    subprocess.Popen(["rasa", "run", "actions"], stdout=log_file, stderr=log_file)
time.sleep(15)  # Give it time to start
print("Action Server is starting in the background. Check action_server.log for details.")
```

```python
# Set ACTION_SERVER_URL for the local environment
os.environ["ACTION_SERVER_URL"] = "http://localhost:5055/webhook"

public_url = ngrok.connect(5005)
print(f"Rasa Public URL: {public_url.public_url}")

print("Starting Rasa Server...")
!rasa run --enable-api --cors "*" --port 5005
```
